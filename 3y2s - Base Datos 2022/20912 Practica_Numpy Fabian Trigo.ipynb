Obten la desviacion estandar de los valores de "mat"

np.std(arreglo, axis = ):
- `axis = None` el arreglo se aplana, convirtiendose en una lista y tomando asi la desviacion
- `axis = 0` toma la desviacion standar dentro de las columnas, (caminando en las filas, osea direccion 0 de la operacion)
- `axis = 1` toma la desviacion standar dentro de cada fila, pero en la direccion axis 1 (direccion columnas)

```python
# puede observar el siguiente ejemplo:
print(mat[0,:])

print( np.std(mat[0,:]) )
print( np.std(mat, axis=1) )
```

```python
std_axisNone = np.std(mat, axis=None)
std_axisZero = np.std(mat, axis=0)
std_axisOne = np.std(mat, axis=1)

print("Desviacion Estandar para:")
print(f"Todos los elementos entre si: {std_axisNone:.3f}")
print(f"Todos los elementos entre si: {std_axisZero}")
print(f"Todos los elementos entre si: {std_axisOne}")
```

### Broadcasting

Dado un arreglo de distancias (en millas) entre las ciudades de la Ruta 66: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff y Los Angeles:



```python
mileposts = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])
```

### Genera un arreglo que incluya todas las distancias llamado distance_array:
- Generar un mileposts traspuesto
- provocar broadcasting entre estos

```python
mileposts[:,np.newaxis] #traspuesta

# La idea es poder restar distancias de manera organizada
# osea si estamos en la 2da ciudad, llevaremos 198 avanzado, por tanto a todo se le restaria 198
# nuestro ultimo destino estaria a 1715 en lugar de 1913 comparado al inicio
distanciasmatrix = mileposts - mileposts[:,np.newaxis]
print(distanciasmatrix)

print('---')
# para eliminar los numeros negativos, que implican volver en el viaje y dejarlo como distancias
distanciasmatrix = np.abs(distanciasmatrix)
print(distanciasmatrix)

# No se toma el cuadrado pues se asume que toda distancia ocurre en la Ruta
# osea solo tiene un grado de lbiertad o una sola dimension 
```

```python
# se puede visualizar:
import matplotlib.pyplot as plt

plt.imshow(distanciasmatrix)
plt.colorbar()
plt.show()
```

Calcula qué distancias se encuentran a menor distancia y genera un plot que permita visualizar el resultado:

- se asume que se refiere a que ciudades mas cercanas entre ellas

```python
distanciasmatrix
```

```python
ordendistancias = np.argpartition(distanciasmatrix, 2, axis=1)
ordendistancias
```

```python
# Y queda graficada la relacion, lo que se pide

plt.imshow(ordendistancias)
plt.colorbar()
plt.show()
```

### que significa lo que ves?
ordenamos los 2 vecinos mas cercanos a cada ubicacion
y lo hicimos manteniendo intactas las filas, pero corriendo en las columnas, de ahi el `axis=1`

Osea que nuestras filas aun mantienen el significado; la fila 0 sera la primera ciudad, la 1 la segunda y asi...

y se nos indica el index de la ciudad mas cercana a ella.
que sera ella misma (osea el index de su propio elemento, y la 2da sera la mas cercana a ella)

observaras que el primer elemento de cada fila es 0,1,2,3,4... hace alucion
al 0 que se encuentra diagonal, osea la distancia consigo mismo




```python
# tomaremos solo las primeras dos columnas
print(ordendistancias[:,:2])
```

### Pero como aplicarlo?

Para hallar las distancias aprovecharemos el arreglo que nos entrega:

```python
# el nombre de las ciudades en orden: (asumire que estan en orden de recorrido, de acuerdo a lo que observo en la imagen)
cadena = "Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles"
ciudades = cadena.split(", ")
```

```python
# index de ciudad mas cercana
indexclosest = ordendistancias[:,1:2]
indexclosest.shape # dimension (10,1) 

# para aprovechar los poderes del indexing, necesitamos reducir la dimension, o flattne
indexclosest = indexclosest.flatten() # [1,2,1,4,3...]
```

```python
# distanciasmatrix[i] selecciona la fila, la ciudad i y sus distancias
# distanciasmatrix[i][j] = distanciasmatrix[i,j] fila i, elemento o columna j

for i, city in enumerate(ciudades):
    j = indexclosest[i]
    distancia = distanciasmatrix[i,j]
    print(f"Para la ciudad {city} la mas cercana esta a {distancia} millas y es {ciudades[j]}")
```
