## Reglas del Broadcasting

Broadcasting en NumPy sigue las siguientes reglas de "interacción" entre arreglos:

- Regla 1: Si los dos arreglos tienen distinto número de dimensiones, el shape del arreglo con menores dimensiones es el que "se rellena" con dimensión uno hacia el lado izquierdo

- Regla 2: Si el shape de los dos arreglos no coinciden en ninguna dimensión, el arreglo con shape 1 en la dimensión correspondiente es el que se "duplica" para llegar a alcanzar el otro shape.

- Regla 3: Si las dimensiones son distintas (los shapes, no el número de dimensiones) y ninguna es "1", entonces se da un error.


Veamos **ejemplos** de las reglas de nuevo

**Ejemplo 1:**

```python
M = np.ones((2, 3))
a = np.arange(3)
print(M)
print(a)
```

Los shapes de los arreglos son

- ``M.shape = (2, 3)``
- ``a.shape = (3,)``

Aplicamos la regla 1: a tiene menos dimensiones, así que "rellenamos" hacia la izquierda con "1":

- ``M.shape -> (2, 3)``
- ``a.shape -> (1, 3)``

Ahora, tenemos el mismo número de dimensiones, pero éstas no son iguales, con lo que "duplicamos" la dimensión del "1":

- ``M.shape -> (2, 3)``
- ``a.shape -> (2, 3)``

## Booleanos en Python
Diferencias entre ocupar "and/or" vs "&/|":

La diferencia es que ``and`` y ``or`` le dicen a python que trate el objeto como una entidad boleana única, mientras que ``&``y ``|`` se refieren a una comparación a nivel de bits

Recordemos que en python, bool() devuelve True para todos los enteros que no sean cero, entonces

```python
bool(2.5), bool(0), bool(True)
```

```python
bool(42 and 0)
```

```python
bool(42 or 0)
```

Pero si ocupamos ``&`` y ``|`` sobre enteros, el "y" y el "o" se aplica a nivel de los bits individuales que componen esos enteros, es decir:

```python
bin(42), bin(59)
```

```python
bin(42 & 59)
```

```python
bin(42 | 59)
```

Ahora, si tenemos un arreglo de boleanos, lo podemos pensar como un conjunto de bits donde 1 es True y 0 es False, y de este modo ``&`` y ``|`` funcionan del modo que esperaríamos:

```python
A = np.array([1, 0, 1, 0, 1, 0], dtype=bool)
B = np.array([1, 1, 1, 0, 1, 1], dtype=bool)
A | B
```

Mientras que ``or`` o ``and`` van a tratar de evaluar el objeto completo, el arreglo completo, y es una operación no definida:

```python
A and B
```

Así que cuando queramos operar con máscaras, como son arreglos de boleanos, lo que debemos de ocupar son ``&`` y ``|``

## SearchSorted Sobre Metodos de Orden
Hemos ocupado ``searchsorted`` pero no hemos hablado de **``sorting``**

¿Cuál es la manera más simple de reordenar un arreglo?

```python
def selection_sort(x):
    """
    
    """
    for i in range(len(x)):
        swap = i + np.argmin(x[i:])
        (x[i], x[swap]) = (x[swap], x[i])
    return x
```

```python
x = np.array([2, 1, 4, 3, 5])
selection_sort(x)
```

Aunque esta manera de ordenar un arreglo sea muy simple, es muy costosa, con el tiempo de cálculo creciendo cuadráticamente con el crecimiento del número de elementos en el arreglo.
$$ \mathcal O (N²)$$

Aunque, por supuesto hay métodos aún menos eficientes:

```python
def bogosort(x):
    """
    Ordena de manera aleatoria, infeciciencia maxima
    """
    while np.any(x[:-1] > x[1:]):
        np.random.shuffle(x)
    return x
```

```python
x = np.array([2, 1, 4, 3, 5])
bogosort(x)
```

Obviamente NumPy tiene métodos más eficientes "pre-cocinados" y esos son los que ocuparemos:

```python
# Ordeno, pero no modifico el arreglo original
x = np.array([2, 1, 4, 3, 5])
print(np.sort(x))
print(x)
```

```python
# Remplazo el arreglo con los valores ordenados
x.sort()
print(x)
```

# K - Nearest Neighbors
Y al igual que existe ``argsort``, existe ``argpartition``. Veamos un ejemplo de cómo estas funciones pueden ser muy útiles en **k-Nearest Neighbors**:

```python
# Generamos un conjunto de 10 puntos random en un espacio bidimensional
rand = np.random.RandomState(42)

X = rand.rand(10, 2)
plt.scatter(X[:, 0], X[:, 1], s=100);
i=0

print("X")
print(X)

print("--- \nDentro del zip:")
for x,y in zip(X[:, 0],X[:, 1]):
    print(x,y)
```

Tengamos el problema de un punto en medio del espacio rodeado de puntos, queremos calcular cual es el punto de dato mas cercano a nuestro punto

Lo que hace K Near es subdividir en secciones o vecindarios puntos que se encuentran vecinos, entonces a la hora de hacer el calculo solo tendremos que calcular dentro de la grilla

```python
# Generamos un conjunto de 10 puntos random en un espacio bidimensional
rand = np.random.RandomState(42)

X = rand.rand(10, 2)
plt.scatter(X[:, 0], X[:, 1], s=100);
i=0
for x,y in zip(X[:, 0],X[:, 1]):
    label = f"({i})"
    i+=1
    plt.annotate(label, # el texto
                 (x,y), # las coordenadas para el texto
                 textcoords="offset points", # donde posicionar el texto (respecto a las coordenadas)
                 xytext=(0,10), # distancia del texto a las coords (x,y)
                 ha='center') # alineamiento horizontal
```

Calculamos ahora la distancia (al cuadrado) entre cada par de puntos ocupando broadcasting y aggregation.


```python
x = X[:, np.newaxis, :]
y = X[np.newaxis, :, :]

print(x)
print("---")
print(y)
print("---")
#print(x-y) # uso esto para hacer un broadcasting
# adquiero asi la diferencia entre cada punto con todos los demas
```

```python
dist_sq = np.sum((X[:, np.newaxis, :] - X[np.newaxis, :, :]) ** 2, axis=-1)
```

Cuando ``axis=-1`` (cualquier número negativo) eso quiere decir que va recorriendo los ejes del último al primero tantas veces como el número diga




```python
dist_sq.shape
```

```python
plt.imshow(dist_sq, cmap="hot");
plt.colorbar();
```

Lo mismo pero paso a paso:

```python
# para cada par de puntos, calculamos la diferencia en las coords
differences = X[:, np.newaxis, :] - X[np.newaxis, :, :]
#differences.shape
differences
```

```python
# elevamos al cuadrado esas diferencias
sq_differences = differences ** 2
#print(sq_differences)
sq_differences.shape
```

```python
# sumamos las diferencias en 'x' y en 'y'
dist_sq = sq_differences.sum(-1)
#print(dist_sq)

dist_sq.shape

plt.imshow(dist_sq);
plt.colorbar();
```

Ahora podemos ocupar argsort para recibir, por fila, los índices de los puntos más cercanos:

```python
nearest = np.argsort(dist_sq, axis=1)
print(nearest)

plt.imshow(nearest);
plt.colorbar();
```

La primera columna son los puntos en sí (cada punto es el más cercano a sí mismo), y la segunda, el punto más cercano, pero ojo, si sólo nos interesa el punto más cercano, o los dos más cercanos, para qué ordenar el resto del arreglo? mejor ocupar argpartition:

```python
K = 2
nearest_partition = np.argpartition(dist_sq, K + 1, axis=1)
```

```python
name = ['Alice', 'Bob', 'Cathy', 'Doug']
age = [25, 45, 37, 19]
weight = [55.0, 85.5, 68.0, 61.5]
```

¿Cómo podemos hacer esto? Dando la estructura desde el ``dtype``(no tiene por qué ser un "simple" ``int``, puede ser mucho más interesante)

```python
data = np.zeros(4, dtype={'names':('name', 'age', 'weight'),
                          'formats':('U10', 'i4', 'f8')})
print(data.dtype)
```

Aqui ``'U10'`` se traduce a "Unicode string of maximum length 10," ``'i4'`` a "4-byte (i.e., 32 bit) integer," y ``'f8'`` a "8-byte (i.e., 64 bit) float."

Una vez definida la estructura, podemos rellenarla con los datos de antes:

```python
data['name'] = name
data['age'] = age
data['weight'] = weight
print(data)
```

```python
# dame todos los nombres
data['name']
```

```python
# Dame la primera fila de datos
data[0]
```

```python
# El nombre de la última fila
data[-1]['name']
```

Y podemos hacer máscaras, por ejemplo: "Nombres de las personas con menos de 30 años"

```python
data[data['age'] < 30]['name']
```

¿Podemos crear estructuras de datos via dtype de otro modo?

```python
#Lo visto antes como diccionario
np.dtype({'names':('name', 'age', 'weight'),
          'formats':('U10', 'i4', 'f8')})
```

```python
#Igual pero especificando los tipos en sí de otro modo:
np.dtype({'names':('name', 'age', 'weight'),
          'formats':((np.str_, 10), int, np.float32)})
```

```python
#Como lista de tuplas en vez de como diccionario:
np.dtype([('name', 'S10'), ('age', 'i4'), ('weight', 'f8')])
```

```python
#Y si no nos importan los nombres de los tipo, i.e. las columnas:
np.dtype('S10,i4,f8')
```

Para los tipos, el primer caracter (opcional) es ``<`` o ``>``, que significa "little endian" o "big endian," y se refiere a la convención ocupada para el orden de bits significativos.
El siguiente caracter especifica el tipo de dato: characters, bytes, ints, floating points, etc (ver tabla).
El último caracter nos dice el tamaño del objeto en bytes.

| Character        | Description           | Example                             |
| ---------        | -----------           | -------                             | 
| ``'b'``          | Byte                  | ``np.dtype('b')``                   |
| ``'i'``          | Signed integer        | ``np.dtype('i4') == np.int32``      |
| ``'u'``          | Unsigned integer      | ``np.dtype('u1') == np.uint8``      |
| ``'f'``          | Floating point        | ``np.dtype('f8') == np.int64``      |
| ``'c'``          | Complex floating point| ``np.dtype('c16') == np.complex128``|
| ``'S'``, ``'a'`` | String                | ``np.dtype('S5')``                  |
| ``'U'``          | Unicode string        | ``np.dtype('U') == np.str_``        |
| ``'V'``          | Raw data (void)       | ``np.dtype('V') == np.void``        |

Uno puede generar tipos compuestos más complicados donde un tipo sea por ejemplo una matriz, es decir:

```python
tp = np.dtype([('id', 'i8'), ('mat', 'f8', (3, 3))])
X = np.zeros(3, dtype=tp)
print(X[0])
print(X['mat'][0])
```
