En el desarrollo del problema 2 del Simulacro de Mayo del 2020 acabamos con:
$2\frac{1- \cos(x)}{x^2}$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): #creamos la funcion
    return(
        (1-np.cos(x))/(x*x)
    )
```

```python
x=np.arange(-1,1,0.01)

plt.grid()
plt.plot(x,f(x))
```

recordaras lo que te hable de tomar varios puntos y si justo tomabas el que contenia la singularidad (que en el caso de arriba sucedio) observaremos un peak

Lo que nos interesa es ver a que valor se aproxima $\frac{1- \cos(x)}{x^2}$; lo hace a 0.5 (eso es lo que es el limite, no el valor exacto que posee alli, eso lo veras en cuando una funcion es continua, si no a que valor parece llegar esto cuando nos movemos a ese punto)

```python
# Ejemplo de lo que sucedia con el limite sin(x)/x para x->0

def g(x): #creamos la funcion
    return(
        (np.sin(x))/(x)
    )
x=np.arange(-0.5,0.5,0.0001)

plt.grid()
plt.plot(x,g(x))
```
